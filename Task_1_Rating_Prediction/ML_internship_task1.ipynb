```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error ,r2_score
```

```python
df = pd.read_csv("Dataset .csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(columns=['Rating color','Rating text'])
```

```python
X = df.drop('Aggregate rating',axis =1)
Y = df['Aggregate rating']
```

```python
X.select_dtypes(include='object').columns
```

```python
X.isnull().sum()
```

```python
drop_cols = ['Restaurant Name','Address','Locality Verbose','Rating color','Rating text']
X = X.drop(columns=drop_cols)
```

```python
X['Cuisines'].isnull().sum()
X= X.dropna()
Y=Y.loc[X.index]
```

```python
binary_cols = ['Has Table booking','Has Online delivery','Is delivering now','Switch to order menu']
for col in binary_cols:
    X[col] = X[col].map({'Yes':1,'No':0})
```

```python
le = LabelEncoder()

cat_cloumns = X.select_dtypes(include='object').columns
for col in cat_cloumns:
    X[col] = le.fit_transform(X[col])
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)

mse = mean_squared_error(Y_test,Y_pred)
r2 = r2_score(Y_test,Y_pred)

print("Mean Squared Error:",mse)
print("R-squared:",r2)
```

```python
feature_importance = pd.DataFrame({'Feature':X.columns,'Coefficient':model.coef_}).sort_values(by='Coefficient',ascending=False)
feature_importance
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(random_state=42)
dt.fit(X_train,Y_train)

dt_pred = dt.predict(X_test)

dt_mse = mean_squared_error(Y_test,dt_pred)
dt_r2 = r2_score(Y_test,dt_pred)

print("Mean Squared Error:",dt_mse)
print("R-squared:",dt_r2)
```
