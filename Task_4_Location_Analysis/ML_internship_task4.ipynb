```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Dataset .csv')
df.head()
```

```python
df[['City','Latitude','Longitude']].isnull().sum()
```

```python
city_counts = df['City'].value_counts().head(10)

plt.figure()
city_counts.plot(kind='bar')
plt.title("Top 10 Cities with the Most Restaurants")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
avg_rating_by_city = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(10)
plt.figure()
avg_rating_by_city.plot(kind='bar')
plt.title("Top 10 Cities by Average Restaurant Rating")
plt.xlabel("City")
plt.ylabel("Average Rating")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='Price range', y='Aggregate rating', data=df)
plt.title("Price Range vs. Average Restaurant Rating")
plt.xlabel("Price Range")
plt.ylabel("Average Rating")
plt.show()
```

```python
plt.figure()
plt.scatter(df['Longitude'],df['Latitude'],alpha=0.4)
plt.title("Geographical Distribution of Restaurants")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
